```python
import spacy
# loading the spaacy English Language
nlp=spacy.load('en_core_web_sm')
```

```python
# Create my text as nlp object
my_text= nlp(u"""With most COVID-19 vaccines, you will need two shots in order for them to work.
 Get the second shot even if you have side effects after the first one, unless a vaccination provider or your doctor tells you not to get a second shot.
It takes time for your body to build protection after any vaccination.
 COVID-19 vaccines that require 2 shots may not protect you until a week or two after your second shot.""")
```

```python
from spacy import displacy
# Render the dependency parse :
displacy.render(my_text, style='dep',jupyter=True)# showing syntactic dependency and POS tags
```

```python
options={'distance':90,'compact':'True','color':'blue','bg':'#F9E79F'}
displacy.render(my_text, style='dep',jupyter=True,options=options)# showing syntactic dependency and POS tags
```

The dependency parse shows the coarse POS tag for each token, as well as the **dependency tag** if given:

```python
for token in my_text:
    print(f'{token.text:{10}} {token.pos_:{7}} {token.dep_:{7}} {spacy.explain(token.dep_)}')
```
