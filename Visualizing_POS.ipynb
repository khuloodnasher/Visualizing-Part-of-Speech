```python
import spacy
# loading the spaacy English Language
nlp=spacy.load('en_core_web_sm')
# Create my text as nlp object
doc1=nlp(u"With most COVID-19 vaccines, you will need two shots in order for them to work.")
```

```python
from spacy import displacy
# Render the dependency parse :
displacy.render(doc1, style='dep',jupyter=True)# showing syntactic dependency and POS tags
```

```python
doc1=nlp(u"With most COVID-19 vaccines, you will need two shots in order for them to work.")
displacy.render(doc1, style='dep',jupyter=True,options={'distance':80})# showing syntactic dependency and POS tags
```
